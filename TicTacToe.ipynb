```python
def winner(game, char):
    #print(game)
    for j in range(len(game)):
        ccnt=0
        for i in range(len(game)):
            #print(game[i][j])
            if game[i][j] == char:
                ccnt +=1
        if ccnt==3 :
            #print("winner wins through column")
            return "Y"
    for j in range(len(game)):
        rcnt=0
        for i in range(len(game)):
            if game[j][i] == char:
                rcnt +=1
        if rcnt==3 :
            #print("winner wins through row")
            return "Y"
        dcnt=0
        cdcnt=0
    for i in range(len(game)):
        #print(game[i][i])
        if game[i][i] == char:
            dcnt +=1
        if game[i][2-i] == char:
            cdcnt +=1
    if dcnt==3 or cdcnt==3:
        #print("winner wins through diagonal")
        return "Y"
    #print(ccnt, rcnt, dcnt)


def updategrid(grid1, position, char):
    #print(position[1])
    if len(position)!=3:
        print("you entered incorrect no. of characters")
        return("NA")
    elif position[1]!= ',' :
        print("you entered without a comma")
        return("NA")
    elif position[0] not in "0123456789" or position[2] not in "0123456789":
        print("you entered a non-digit")
        return("NA")
    else :
        spots=position.split(',')
        row=int(spots[0])
        column=int(spots[1])
        if (1<=row<=3) and (1<=column<=3):
            if grid1[row-1][column-1]==' ':
                grid1[row-1][column-1]=char
                return(grid1)
            else :
                print(char + " position taken")
                return(grid1)
        else :
            print("values need to be between 1 and 3 inclusive row was {} column was {}".format(row,column))
            return("NA")
flag="y"
charx="X"
charo="O"
grid = [[' ',' ',' '],
        [' ',' ',' '],
        [' ',' ',' ']]
print("This is the game of Tic-Tac-Toe for 3x3 grid. Each player takes turn to either enter X or O.")
print("Whoever, first forms a straight line wins.")
print("Please enter positions in the form 1,3 for row=1 and column=3.")

while flag=="y":
    position1=input("Player 1 enter a position for X: ")
    grid=updategrid(grid, position1, charx)
    if grid=="NA" : break
    result_player1=winner(grid, charx)
    if result_player1 == "Y" :
        print("Player 1 wins")
        flag="n"
        break
    position2=input("Player 2 enter a position for O: ")
    grid=updategrid(grid, position2, charo)
    result_player2=winner(grid, charo)
    if result_player2 == "Y" :
        print("Player 2 wins")
        flag="n"

while flag=="n" :
    print(grid[0])
    print(grid[1])
    print(grid[2])
    break
```
