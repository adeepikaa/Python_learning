```python
# This code is to play the game of Hangman!
# The way to play it is, a random word is chosen by the program and the user
# has to guess the word by starting with each letter one at a time.
# When a letter is guessed correctly to be in a word the program will
# list all locations of that letter.

# libraries for random pick of the word and string manipulations
import random
import string

# sowpods.txt is the txt file which has a list of pre-gathered words that can be used for the game.
sowpodsfile=open("sowpods.txt", 'r')
sowpod=[]

#reading all words into a variable to randomly pick a word
for i in sowpodsfile:
    sowpod.append(i)

#random pick of workd
word=random.choice(sowpod).strip()
#print(word)
sowpodsfile.close()


print("Welcome to HANGMAN!")

# initialize defaults
flag=1
cnt=0
guess_word=''
guesslist=[]

# fill word with ? marks 1st
for i in range(len(word)):
    guess_word=guess_word+'?'
print(guess_word)

char=input("Guess a letter : ")

# use the read character and check if its in randomly selected word
while flag:
    guessword=list(guess_word)
    
    # Check for different faulty entries
    if char not in string.ascii_letters:
        print("Enter only letters")
        
    elif len(char)!=1 :
        print("Enter 1 letter")
        
    elif char in guessword :
        print("Already guessed the letter")
        
    else :
        # counter to count number of tries
        cnt+=1
        # change to uppercase
        for i in range(len(word)) :
            # check if character is in the word(full length of word)
            # if yes, replace ? with new letter
            if char.upper()==word[i] :
                guessword[i]=char.upper()
            else:
                # else append guess list 
                if char.upper() not in guesslist:
                    guesslist.append(char.upper())
        # join each letter or ? mark back into word
        guess_word=''.join(guessword)
    # check if word matched print results
    if word==guess_word:
        print("You got it in {} attempts".format(cnt))
        print(guess_word)
        flag=0
    else:
        print(guess_word)
        char=input("Guess a letter other than already entered {} :".format(' '.join(guesslist)))
```
